```python
# how to create a parabola:
x= np.arange(-9,10)
y= x**2
plt.plot(x,y)
```

```python
# labelling the axis:
plt.plot(x,y)
plt.xlabel("x values")
plt.ylabel("y vabels")
plt.title("parabola")
plt.xticks(np.arange(-9,10,4))
plt.yticks(y)
```

```python
plt.plot(x,y)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("parabola")
plt.xlim(-5,50) # limit to display the x values.
plt.ylim(10,100) # limits to display the y values.
```

```python
# to darw a multiiple graph
x=np.arange(10,20)

plt.subplot(2,2,1)
plt.plot(x)
plt.title("graph 1")

plt.subplot(2,2,2)
plt.plot(x,x+3)
plt.title("graph 2")

plt.subplot(2,2,3)
plt.plot(x,x*2)
plt.title("graph 3")

plt.subplot(2,2,4)
plt.plot(x,np.log(x))
plt.title("graph 4")
```

```python
# to darw a multiiple graph
x=np.arange(10,20)

plt.subplot(2,2,1,frameon=False) # by default the frameon is true 
                             #if we make it as false then there wil be no frame

plt.plot(x)
plt.title("graph 1")

plt.subplot(2,2,2)
plt.plot(x,x+3)
plt.title("graph 2")

plt.subplot(2,2,3)
plt.plot(x,x*2)
plt.title("graph 3")

plt.subplot(2,2,4)
plt.plot(x,np.log(x))
plt.title("graph 4")
```
